### Question 1: Mars Transfer Orbit

The transfer orbit from Earth to Mars will have a semi-major axis of 

$a = \frac{r_e + r_m}{2}$

$r_e = 1 AU$, and $r_m = 1.524 AU$, so

$a = \frac{2.524}{2} = 1.262$

$ $

From Kepler's third law: $T^2 = a^3$

$T = \sqrt{a^3} = 1.262^{\frac{3}{2}} = 1.4177 years$

The voyage will take half of the time of a period, so the the voyage will take $\frac{1.4177}{2} = 0.70885 years$, or about 258.9 days.

$ $

In order to enter the elliptical orbit from orbiting around the earth, there must be an impulse of

$\Delta v_1 = \sqrt{\frac{\mu}{r_1}}(\sqrt{\frac{2r_2}{r_1 + r_2}} - 1)$

$\Delta v_1 = \sqrt{\frac{GM_☉}{1 AU}}(\sqrt{\frac{2 * 1.524 AU}{2.524 AU}} - 1)$

$\Delta v_1 = \sqrt{\frac{GM_☉}{1 AU}}(\sqrt{\frac{2 * 1.524 AU}{2.524 AU}} - 1)$

$\Delta v_1 = \sqrt{\frac{GM_☉}{1 AU}}(0.099)$

$\Delta v_1 = \sqrt{\frac{4.30091 * 10^{−3} (pc * M_☉^{-1} * (\frac{km}{s})^2) * 1 M_☉}{4.84814 * 10^{-6} pc}}
(0.099)$

$\Delta v_1 = \sqrt{\frac{4.30091 * 10^{−3} }{4.84814 * 10^{-6}}}(0.099)(\frac{km}{s}))$

$\Delta v_1 = \sqrt{0.88712 * 10^3}(0.099)(\frac{km}{s}))$

$\Delta v_1 = \sqrt{0.88712 * 10^3}(0.099)(\frac{km}{s}))$

$\Delta v_1 = 2.94 \frac{km}{s}$

$ $

To escape the elliptical orbit and continue along with the orbit at mars's distance, there must be an impulse of

$\Delta v_2 = \sqrt{\frac{\mu}{r_2}}(1 - \sqrt{\frac{2r_1}{r_1 + r_2}})$

$\Delta v_2 = \sqrt{\frac{GM_☉}{1.524 AU}}(1 - \sqrt{\frac{2 * 1 AU}{2.524 AU}})$

$\Delta v_2 = \sqrt{\frac{4.30091 * 10^{−3} (pc * M_☉^{-1} * (\frac{km}{s})^2) * 1 M_☉}{7.388561 * 10^{-6} pc}}(0.1098)$

$\Delta v_2 = \sqrt{\frac{4.30091 * 10^{−3}}{7.388561 * 10^{-6}}}(0.1098)\frac{km}{s}$

$\Delta v_2 = \sqrt{0.5821 * 10^3}(0.1098)\frac{km}{s}$

$\Delta v_2 = 2.6491 \frac{km}{s}$

$ $



$ $
In order to transfer from orbiting the Sun at Eath to orbiting the Sun at Mars, we take advantage of gravitational potential energy.  Flying against the pull of gravity, even when Earth and Mars are closest, takes a lot of fuel.  Therefore, we launch with an impulse of 2.94 km/s to exit earth's typical orbit around the sun, and enter an elliptical orbit.  We therefore let the gravitational force of the sun bring us to the distance of Mars's orbit. From there, we have a second impulse of 2.6491 km/s to take us out of the elliptical orbit and into an orbit of the sun at Mars's distance from the Sun.
